### Quali sono i Pokémon leggendari? 

```python
leggendari=pokemon.loc[:, 'Legendary']==True      
pokemon.loc[leggendari]  
```

### E quali sono i leggendari di tipo 1 Grass?

```python
leggendari=pokemon.loc[:, 'Legendary']==True     
tipo1_grass=pokemon.loc[:, 'Type 1']== 'Grass'

filtro= leggendari & tipo1_grass
pokemon.loc[filtro]  
```

### E leggendari di tipo 1 Ice o Fire? 

```python
leggendari=pokemon.loc[:, 'Legendary']==True     
tipo1_ice=pokemon.loc[:, 'Type 1']== 'Ice'
tipo1_fire=pokemon.loc[:, 'Type 1']== 'Fire'
filtro_tipo = tipo1_fire | tipo1_ice
filtro= filtro_tipo & leggendari
pokemon.loc[filtro]
```

### Quali sono le medie di charges rispetto a region? Ci sono differenze significative?
